## Manipulating Columns

### rename column

```python
nba = df.copy()
nba.head(2)
```

```python
nba.rename(columns={'game_location':'loc', 'game_result':'result'}, inplace=True)
```

```python
nba.head()
```

### drop columns

```python
nba = nba.drop(labels=['elo_i', 'elo_n'], axis=1)
```

```python
nba.head()
```

### creating column

```python
nba['selisih_poin'] = nba['pts'] - nba['opp_pts']
nba.head()
```

```python
# siapkan list kosong, sebagai tempat penampung
# specify variabel yang dibutuhkan
# buat prosedur -> hasil
# hasilnya simpan ke list kosong
# loop selesai, list penuh -> kolom baru

hasil = []

# kolom pts
# kolom opp_pts

for i in range(0, len(nba)): 
    selisih = nba.loc[i, 'pts'] - nba.loc[i, 'opp_pts']
    hasil.append(selisih)

nba['selisih_loop'] = hasil
```

```python
nba.head()
```

```python
# kolom baru hasil pertandingan
# def dunction
def hasil(selisih):
    if selisih > 0:
        return 'menang'
    elif selisih < 0:
        return 'kalah'
    else:
        return 'draw'
```

```python
hasil_pertandingan = []

for i in range(0, len(nba)):
    result = hasil(nba.iloc[i,-2])
    hasil_pertandingan.append(result)

nba['hasil'] = hasil_pertandingan
```
