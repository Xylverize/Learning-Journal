```python
#import library
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('fifa2022.csv')
df.head()
```

```python
df.describe()
```

```python
#buat histogram outlier dengan no outlier
plt.figure(figsize=(20, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Rating'])
plt.title('Data tanpa Outlier Penumpang kecelakaan')
plt.xlabel('Total Orang Dalam Kendaraan')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(x=df['Rating'], bins=50, kde=True)
plt.title('Data Outlier Penumpang kecelakaan')
plt.xlabel('Total Orang Dalam Kendaraan')
plt.ylabel('Frequency')
plt.show()
```
